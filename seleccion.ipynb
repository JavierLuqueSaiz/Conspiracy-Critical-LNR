```python
import pickle
import pandas as pd
```

```python
with open('results', 'rb') as f:
    df_en, df_es = pickle.load(f)

with open('results2.pkl', 'rb') as f:
    df_en2, df_es2 = pickle.load(f)
    
```

```python
df_en.loc["Binary Bag of Words"] = df_en2.loc["Binary Bag of Words"]
df_es.loc["Binary Bag of Words"] = df_es2.loc["Binary Bag of Words"]
df_en.loc["Bag of Words"] = df_en2.loc["Bag of Words"]
df_es.loc["Bag of Words"] = df_es2.loc["Bag of Words"]
df_en.loc["Bigrams of Words"] = df_en2.loc["Bigrams of Words"]
df_es.loc["Bigrams of Words"] = df_es2.loc["Bigrams of Words"]
df_en.dropna(axis=0, inplace=True)
df_es.dropna(axis=0, inplace=True)
```

```python
english = pd.DataFrame(df_en.stack().reset_index().sort_values(by=0, ascending=False))
english.columns = ["Embedding", "Model", "MCC"]
english.head(30)
```

```python
espanol = pd.DataFrame(df_es.stack().reset_index().sort_values(by=0, ascending=False))
espanol.columns = ["Embedding", "Model", "MCC"]
espanol.head(30)
```
